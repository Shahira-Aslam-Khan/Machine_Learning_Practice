```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import cv2
```

```python
stars = np.array(Image.open('/stars.jpg').convert('RGB')).astype(np.uint8)
```

```python
size=stars.shape
```

```python
stars = stars.reshape(-1,3)
```

```python
Kmean=KMeans(n_clusters=2)
```

```python
Kmean.fit(stars)
```

```python
clusters=Kmean.labels_
clustered_stars=clusters.reshape(size[0],size[1]).astype('uint8')
plt.figure(figsize=(15,10))
plt.imshow(clustered_stars,cmap='gray')
plt.axis('off')
```

```python
plt.show()
```

```python
print("There are %u stars in the sky"% cv2.connectedComponents(clustered_stars)[0])
```
